# Fonctions communes — PolyBERT
Ce fichier charge le modèle et définit toutes les fonctions réutilisables.
Il est destiné à être exécuté via `%run polybert_functions.ipynb` dans les autres scripts.

```python
import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import wikipediaapi
import torch
from transformers import AutoTokenizer, AutoModel
from scipy.spatial.distance import cosine
from itertools import combinations
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from scipy.stats import spearmanr, pearsonr
from matplotlib.patches import Patch
import nltk
from nltk.corpus import wordnet as wn
nltk.download('wordnet', quiet=True)
nltk.download('omw-1.4', quiet=True)
```

```python
MODEL_NAME = "camembert-base"
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
model = AutoModel.from_pretrained(MODEL_NAME)
model.eval()
device = torch.device("cpu")
model = model.to(device)

wiki = wikipediaapi.Wikipedia(language='fr', user_agent='projet-polysemie/1.0')
print(f"Modèle chargé : {MODEL_NAME}")
```

```python
def extraire_phrases(texte, mot, n_max=30):
    phrases = re.split(r'(?<=[.!?])\s+', texte)
    pattern = r'\b' + re.escape(mot) + r'\b'
    resultats = [
        p.strip() for p in phrases
        if re.search(pattern, p, re.IGNORECASE) and 20 < len(p) < 300
    ]
    return resultats[:n_max]
```

```python
def get_word_embedding(phrase, mot):
    inputs = tokenizer(phrase, return_tensors="pt", truncation=True, max_length=512).to(device)
    tokens = tokenizer.convert_ids_to_tokens(inputs["input_ids"][0])
    mot_lower = mot.lower()
    indices = []
    i = 0
    while i < len(tokens):
        token_propre = tokens[i].replace("\u2581", "").lower()
        mot_reconstruit = token_propre
        j = i + 1
        while j < len(tokens):
            if tokens[j].startswith("\u2581"):
                break
            mot_reconstruit += tokens[j].replace("\u2581", "").lower()
            j += 1
        if mot_reconstruit == mot_lower:
            indices = list(range(i, j))
            break
        i += 1
    if not indices:
        return None
    with torch.no_grad():
        outputs = model(**inputs)
    vecteur = outputs.last_hidden_state[0][indices].mean(dim=0)
    return vecteur.cpu().numpy()
```

```python
def build_corpus(articles_par_mot):
    """Construit le corpus depuis un dict {mot: [titres Wikipedia]}."""
    corpus = {}
    for mot, articles in articles_par_mot.items():
        phrases_mot = []
        for titre in articles:
            page = wiki.page(titre)
            if page.exists():
                phrases_mot += extraire_phrases(page.text, mot)
        corpus[mot] = list(set(phrases_mot))
        print(f"{mot:20s}: {len(corpus[mot])} phrases")
    return corpus
```

```python
def build_corpus_auto(mots, n_max=20):
    """Construit le corpus depuis une liste de mots en cherchant automatiquement sur Wikipedia."""
    corpus = {}
    for mot in mots:
        page = wiki.page(mot.capitalize())
        phrases = extraire_phrases(page.text, mot, n_max=n_max) if page.exists() else []
        corpus[mot] = phrases
        if phrases:
            print(f"{mot:20s}: {len(phrases)} phrases")
    return corpus
```

```python
def extract_embeddings(corpus):
    """Extrait les embeddings CamemBERT pour chaque mot du corpus."""
    embeddings = {}
    phrases_retenues = {}
    for mot, phrases in corpus.items():
        vecteurs, ok = [], []
        for phrase in phrases:
            vec = get_word_embedding(phrase, mot)
            if vec is not None:
                vecteurs.append(vec)
                ok.append(phrase)
        embeddings[mot] = vecteurs
        phrases_retenues[mot] = ok
        if vecteurs:
            print(f"{mot:20s}: {len(vecteurs)} embeddings")
    return embeddings, phrases_retenues
```

```python
def score_polysemie(vecteurs, n_max=50):
    """Score = écart-type des similarités cosinus 2-à-2. Plus élevé = plus polysémique."""
    vecs = vecteurs[:n_max]
    sims = np.array([1 - cosine(v1, v2) for v1, v2 in combinations(vecs, 2)])
    return {"score": np.std(sims), "sim_moy": np.mean(sims), "n_paires": len(sims)}
```

```python
def compute_scores_df(embeddings, mots_polysemique=[]):
    """Calcule les scores de polysémie et retourne un DataFrame trié."""
    rows = []
    for mot, vecs in embeddings.items():
        if len(vecs) < 2:
            continue
        res = score_polysemie(vecs)
        type_mot = "polysémique" if mot in mots_polysemique else ("monosémique" if mots_polysemique else "?")
        rows.append({
            "mot": mot,
            "score_polysemie": res["score"],
            "sim_moyenne": res["sim_moy"],
            "n_embeddings": len(vecs),
            "type": type_mot
        })
    return pd.DataFrame(rows).sort_values("score_polysemie", ascending=False).reset_index(drop=True)
```

```python
def viz_bar_scores(df, titre="Score de polysémie", fichier=None):
    palette = {"polysémique": "red", "monosémique": "green", "?": "steelblue"}
    couleurs = df["type"].map(palette)
    fig, ax = plt.subplots(figsize=(12, max(5, len(df) * 0.35)))
    ax.barh(df["mot"], df["score_polysemie"], color=couleurs)
    ax.axvline(df["score_polysemie"].mean(), color="gray", linestyle="--", label="moyenne")
    ax.set_xlabel("Score de polysémie (écart-type cosinus)")
    ax.set_title(titre)
    handles = [Patch(color=c, label=l) for l, c in palette.items() if l in df["type"].values]
    handles.append(Patch(color="gray", label="moyenne"))
    ax.legend(handles=handles)
    plt.tight_layout()
    if fichier:
        plt.savefig(fichier, dpi=150)
    plt.show()
```

```python
def viz_mot_pca(mot, embeddings, phrases_retenues, mots_polysemique=[], n_labels=3):
    vecs = np.array(embeddings[mot])
    coords = PCA(n_components=2).fit_transform(vecs)
    couleur = "red" if mot in mots_polysemique else "green"
    fig, ax = plt.subplots(figsize=(6, 5))
    ax.scatter(coords[:, 0], coords[:, 1], alpha=0.7, color=couleur, s=60)
    for i in range(min(n_labels, len(coords))):
        ax.annotate(phrases_retenues[mot][i][:40] + "...", coords[i], fontsize=6, alpha=0.6)
    ax.set_title(f"\u00ab {mot} \u00bb \u2014 PCA 2D")
    ax.set_xlabel("PC1")
    ax.set_ylabel("PC2")
    plt.tight_layout()
    plt.savefig(f"viz_{mot}.png", dpi=150)
    plt.show()
```

```python
def viz_tsne_global(embeddings, mots_polysemique=[], fichier="tsne_global.png"):
    all_vecs, all_mots = [], []
    for mot, vecs in embeddings.items():
        for vec in vecs:
            all_vecs.append(vec)
            all_mots.append(mot)
    X = np.array(all_vecs)
    X_tsne = TSNE(n_components=2, perplexity=30, random_state=42, max_iter=1000).fit_transform(X)
    mots_uniques = list(embeddings.keys())
    palette = plt.cm.tab20.colors
    couleur_par_mot = {m: palette[i % 20] for i, m in enumerate(mots_uniques)}
    fig, ax = plt.subplots(figsize=(14, 9))
    for mot in mots_uniques:
        idx = [i for i, m in enumerate(all_mots) if m == mot]
        coords = X_tsne[idx]
        marker = "o" if mot in mots_polysemique else "^"
        ax.scatter(coords[:, 0], coords[:, 1], color=couleur_par_mot[mot],
                   marker=marker, alpha=0.7, s=40, label=mot)
        ax.annotate(mot, (coords[:, 0].mean(), coords[:, 1].mean()),
                    fontsize=8, fontweight="bold", color=couleur_par_mot[mot],
                    bbox=dict(boxstyle="round,pad=0.2", fc="white", alpha=0.6))
    ax.set_title("t-SNE \u2014 Embeddings CamemBERT (\u25cf poly\u00e9mique  \u25b2 monosmique)")
    ax.legend(bbox_to_anchor=(1.01, 1), loc="upper left", fontsize=7, ncol=1, framealpha=0.7)
    plt.tight_layout()
    plt.savefig(fichier, dpi=150, bbox_inches="tight")
    plt.show()
```

```python
def nb_sens_wordnet(mot, lang='fra'):
    return len(wn.synsets(mot, lang=lang))
```

```python
def viz_correlation_wordnet(df, fichier="correlation_wordnet.png"):
    df = df.copy()
    df["nb_sens"] = df["mot"].apply(nb_sens_wordnet)
    df = df[df["nb_sens"] > 0].copy()
    rho, p_rho = spearmanr(df["score_polysemie"], df["nb_sens"])
    r, p_r = pearsonr(df["score_polysemie"], df["nb_sens"])
    print(f"Spearman \u03c1={rho:.3f} (p={p_rho:.4f})  |  Pearson r={r:.3f} (p={p_r:.4f})")
    palette = {"polysémique": "orange", "monosémique": "green", "?": "steelblue"}
    couleurs = df["type"].map(palette)
    fig, ax = plt.subplots(figsize=(9, 6))
    ax.scatter(df["nb_sens"], df["score_polysemie"], c=couleurs, s=100, alpha=0.8, zorder=3)
    for _, row in df.iterrows():
        ax.annotate(row["mot"], (row["nb_sens"], row["score_polysemie"]),
                    xytext=(6, 4), textcoords="offset points", fontsize=8)
    m, b = np.polyfit(df["nb_sens"], df["score_polysemie"], 1)
    x_line = np.linspace(df["nb_sens"].min(), df["nb_sens"].max(), 100)
    ax.plot(x_line, m * x_line + b, color="gray", linestyle="--")
    ax.set_xlabel("Nombre de sens (WordNet)")
    ax.set_ylabel("Score de polysémie BERT")
    ax.set_title(f"Corrélation BERT vs WordNet\nSpearman \u03c1={rho:.3f}, p={p_rho:.4f}")
    handles = [Patch(color=c, label=l) for l, c in palette.items() if l in df["type"].values]
    ax.legend(handles=handles)
    plt.tight_layout()
    plt.savefig(fichier, dpi=150)
    plt.show()
    return df
```
