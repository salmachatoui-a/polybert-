# Étude de la polysémie — CamemBERT (FR) vs BERT (EN)


```python
# ============================================================
# CELLULE 1 — CONFIGURATION : mots et articles Wikipedia
# ============================================================

# ---------- FRANÇAIS ----------
mots_poly_fr  = ["café", "souris", "porte", "table", "avocat",
                 "tour", "bois", "livre", "vol", "chat"]
mots_mono_fr  = ["médecin", "volcan", "escalier", "atome", "chaise"]

articles_fr = {
    "café":     ["Café", "Café (établissement)"],
    "souris":   ["Souris", "Souris (informatique)"],
    "porte":    ["Porte", "Porte (architecture)"],
    "table":    ["Table de multiplication", "Table des matières"],
    "avocat":   ["Avocat (fruit)", "Avocat (métier)"],
    "tour":     ["Tour (architecture)", "Tour (machine-outil)",
                 "Tour de France", "Tour de magie", "Tournée"],
    "bois":     ["Bois", "Bois (matériau)"],
    "livre":    ["Livre (document)", "Livre (unité de masse)", "Livre sterling"],
    "vol":      ["Vol (aviation)", "Vol (droit)", "Migration des oiseaux"],
    "chat":     ["Chat", "Chat (informatique)"],
    "médecin":  ["Médecin"],
    "volcan":   ["Volcan", "Éruption volcanique"],
    "escalier": ["Escalier"],
    "atome":    ["Atome"],
    "chaise":   ["Chaise"],
}

nb_sens_fr = {
    "café": 5, "souris": 3, "porte": 6, "table": 7, "avocat": 3,
    "tour": 8, "bois": 5, "livre": 4, "vol": 5, "chat": 4,
    "médecin": 1, "volcan": 1, "escalier": 1, "atome": 2, "chaise": 2,
}

# ---------- ANGLAIS ----------
mots_poly_en  = ["mouse", "bank", "bat", "bark", "crane",
                 "light", "ring", "file", "pitcher", "book"]
mots_mono_en  = ["doctor", "volcano", "staircase", "atom", "chair"]

articles_en = {
    "mouse":     ["Mouse", "Computer mouse"],
    "bank":      ["Bank", "River bank", "Bank (geography)"],
    "bat":       ["Bat", "Cricket bat", "Baseball bat"],
    "bark":      ["Bark (botany)", "Bark (sound)"],
    "crane":     ["Crane (bird)", "Crane (machine)"],
    "light":     ["Light", "Lighting", "Light (physics)"],
    "ring":      ["Ring (jewellery)", "Boxing ring", "Telephone call"],
    "file":      ["Computer file", "File (tool)"],
    "pitcher":   ["Pitcher (baseball)", "Pitcher (container)"],
    "book":      ["Book", "Bookmaker", "Booking"],
    "doctor":    ["Physician"],
    "volcano":   ["Volcano", "Volcanic eruption"],
    "staircase": ["Staircase"],
    "atom":      ["Atom"],
    "chair":     ["Chair"],
}

nb_sens_en = {
    "mouse": 3, "bank": 9, "bat": 4, "bark": 4, "crane": 3,
    "light": 10, "ring": 7, "file": 5, "pitcher": 3, "book": 6,
    "doctor": 1, "volcano": 1, "staircase": 1, "atom": 2, "chair": 3,
}

print('✅ Configuration chargée')
print(f'FR : {len(mots_poly_fr)} polysémiques + {len(mots_mono_fr)} monosémiques')
print(f'EN : {len(mots_poly_en)} polysémiques + {len(mots_mono_en)} monosémiques')
```

```python
# ============================================================
# CELLULE 2 — IMPORTS
# ============================================================
import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import wikipediaapi
import torch
from transformers import AutoTokenizer, AutoModel
from scipy.spatial.distance import cosine
from scipy.stats import spearmanr
from itertools import combinations
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import umap
import warnings
warnings.filterwarnings('ignore')

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'Device : {device}')
```

```python
# ============================================================
# CELLULE 3 — FONCTIONS RÉUTILISABLES
# ============================================================

def extraire_phrases(texte, mot, n_max=30):
    phrases = re.split(r'(?<=[.!?])\s+', texte)
    pattern = r'\b' + re.escape(mot) + r'\b'
    resultats = [p.strip() for p in phrases
                 if re.search(pattern, p, re.IGNORECASE)
                 and 20 < len(p) < 300]
    return resultats[:n_max]


def construire_corpus(articles_par_mot, langue='fr'):
    """Extrait les phrases depuis Wikipedia pour une langue donnée."""
    wiki = wikipediaapi.Wikipedia(
        language=langue,
        user_agent='projet-polysemie/1.0'
    )
    corpus = {}
    for mot, articles in articles_par_mot.items():
        phrases_mot = []
        for titre in articles:
            page = wiki.page(titre)
            if page.exists():
                phrases_mot += extraire_phrases(page.text, mot)
            else:
                print(f'  ⚠️  [{langue.upper()}] introuvable : {titre}')
        corpus[mot] = list(set(phrases_mot))
        print(f'  [{langue.upper()}] {mot:15s} : {len(corpus[mot])} phrases')
    return corpus


def get_word_embedding(phrase, mot, tokenizer, model, device, langue='fr'):
    """Retourne le vecteur contextuel du mot dans la phrase.
    Gère CamemBERT (▁) et BERT-base (##).
    """
    inputs = tokenizer(phrase, return_tensors='pt',
                       truncation=True, max_length=512).to(device)
    tokens = tokenizer.convert_ids_to_tokens(inputs['input_ids'][0])
    mot_lower = mot.lower()
    indices = []

    if langue == 'fr':  # CamemBERT : sous-tokens avec ▁
        i = 0
        while i < len(tokens):
            fragment = ''
            j = i
            while j < len(tokens) and len(fragment) <= len(mot_lower) + 3:
                token_clean = tokens[j].replace('▁', '').lower()
                fragment += token_clean
                if fragment == mot_lower:
                    indices = list(range(i, j + 1))
                    break
                j += 1
            if indices:
                break
            i += 1
    else:  # BERT-base : sous-tokens avec ##
        i = 0
        while i < len(tokens):
            if tokens[i].startswith('##'):
                i += 1
                continue
            fragment = tokens[i].lower()
            j = i + 1
            while j < len(tokens) and tokens[j].startswith('##'):
                fragment += tokens[j][2:].lower()
                j += 1
            if fragment == mot_lower:
                indices = list(range(i, j))
                break
            i += 1

    if not indices:
        return None

    with torch.no_grad():
        outputs = model(**inputs)
    hidden_states = outputs.last_hidden_state[0]
    embedding = hidden_states[indices].mean(dim=0).cpu().numpy()
    return embedding


def extraire_embeddings(corpus, mots_poly, tokenizer, model, device, langue='fr'):
    """Extrait les embeddings pour tous les mots du corpus."""
    embeddings = {}
    phrases_retenues = {}
    for mot, phrases in corpus.items():
        vecs, phrases_ok = [], []
        for phrase in phrases:
            vec = get_word_embedding(phrase, mot, tokenizer, model, device, langue)
            if vec is not None:
                vecs.append(vec)
                phrases_ok.append(phrase)
        embeddings[mot] = vecs
        phrases_retenues[mot] = phrases_ok
        statut = '✅' if len(vecs) >= 10 else '⚠️ '
        print(f'  {statut} [{langue.upper()}] {mot:15s} : {len(vecs)} embeddings')
    return embeddings, phrases_retenues


def score_polysemie(vecteurs, n_max=50):
    vecs = vecteurs[:n_max]
    sims = [1 - cosine(v1, v2) for v1, v2 in combinations(vecs, 2)]
    sims = np.array(sims)
    return {'score': np.std(sims), 'sim_moy': np.mean(sims),
            'sim_min': np.min(sims), 'n_paires': len(sims)}


def calculer_scores(embeddings, mots_poly, nb_sens):
    """Calcule les scores et retourne un DataFrame."""
    rows = []
    for mot, vecs in embeddings.items():
        if len(vecs) < 5:
            continue
        r = score_polysemie(vecs)
        rows.append({
            'mot': mot,
            'score_bert': r['score'],
            'sim_moy': r['sim_moy'],
            'nb_sens': nb_sens.get(mot, 0),
            'type': 'polysémique' if mot in mots_poly else 'monosémique'
        })
    return pd.DataFrame(rows).sort_values('score_bert', ascending=False)

print('✅ Fonctions chargées')
```

```python
# ============================================================
# CELLULE 4 — CORPUS FRANÇAIS
# ============================================================
print('📖 Extraction corpus FR...')
corpus_fr = construire_corpus(articles_fr, langue='fr')
```

```python
# ============================================================
# CELLULE 5 — CORPUS ANGLAIS
# ============================================================
print('📖 Extraction corpus EN...')
corpus_en = construire_corpus(articles_en, langue='en')
```

```python
# ============================================================
# CELLULE 6 — CHARGEMENT MODÈLES
# ============================================================

# Modèle français
print('🔄 Chargement CamemBERT (FR)...')
tokenizer_fr = AutoTokenizer.from_pretrained('camembert-base')
model_fr     = AutoModel.from_pretrained('camembert-base').eval().to(device)
print('✅ CamemBERT prêt')

# Modèle anglais
print('🔄 Chargement BERT-base (EN)...')
tokenizer_en = AutoTokenizer.from_pretrained('bert-base-uncased')
model_en     = AutoModel.from_pretrained('bert-base-uncased').eval().to(device)
print('✅ BERT-base prêt')
```

```python
# ============================================================
# CELLULE 7 — EMBEDDINGS FRANÇAIS
# ============================================================
print('🧠 Extraction embeddings FR...')
embeddings_fr, phrases_fr = extraire_embeddings(
    corpus_fr, mots_poly_fr, tokenizer_fr, model_fr, device, langue='fr'
)
```

```python
# ============================================================
# CELLULE 8 — EMBEDDINGS ANGLAIS
# ============================================================
print('🧠 Extraction embeddings EN...')
embeddings_en, phrases_en = extraire_embeddings(
    corpus_en, mots_poly_en, tokenizer_en, model_en, device, langue='en'
)
```

```python
# ============================================================
# CELLULE 9 — SCORES DE POLYSÉMIE
# ============================================================
df_fr = calculer_scores(embeddings_fr, mots_poly_fr, nb_sens_fr)
df_en = calculer_scores(embeddings_en, mots_poly_en, nb_sens_en)

print('\n=== SCORES FRANÇAIS ===')
print(df_fr[['mot','score_bert','nb_sens','type']].to_string(index=False))
print('\n=== SCORES ANGLAIS ===')
print(df_en[['mot','score_bert','nb_sens','type']].to_string(index=False))
```

```python
# ============================================================
# CELLULE 10 — VISUALISATION : scores côte à côte
# ============================================================
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

for ax, df, titre in zip(axes,
                         [df_fr, df_en],
                         ['CamemBERT — Français', 'BERT-base — Anglais']):
    df_plot = df.sort_values('score_bert')
    couleurs = df_plot['type'].map({'polysémique': '#E85D24', 'monosémique': '#1D9E75'})
    ax.barh(df_plot['mot'], df_plot['score_bert'], color=couleurs)
    ax.axvline(df_plot['score_bert'].mean(), color='gray',
               linestyle='--', linewidth=1, label='moyenne')
    ax.set_xlabel('Score de polysémie (écart-type cosinus)')
    ax.set_title(titre)
    ax.legend(handles=[
        mpatches.Patch(color='#E85D24', label='polysémique'),
        mpatches.Patch(color='#1D9E75', label='monosémique'),
        mpatches.Patch(color='gray',    label='moyenne')
    ])

plt.tight_layout()
plt.savefig('scores_fr_en.png', dpi=150)
plt.show()
print('✅ Sauvegardé : scores_fr_en.png')
```

```python
# ============================================================
# CELLULE 11 — CORRÉLATION WORDNET : FR et EN
# ============================================================
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

for ax, df, titre, lang in zip(
    axes,
    [df_fr, df_en],
    ['Français (CamemBERT)', 'Anglais (BERT-base)'],
    ['fr', 'en']
):
    x = df['nb_sens'].values
    y = df['score_bert'].values
    rho, p = spearmanr(x, y)

    couleurs = df['type'].map({'polysémique': '#E85D24', 'monosémique': '#1D9E75'})
    ax.scatter(x, y, c=couleurs, s=80, alpha=0.85, zorder=3)

    for _, row in df.iterrows():
        ax.annotate(row['mot'], (row['nb_sens'], row['score_bert']),
                    textcoords='offset points', xytext=(5, 3), fontsize=7)

    m, b = np.polyfit(x, y, 1)
    x_line = np.linspace(x.min(), x.max(), 100)
    ax.plot(x_line, m * x_line + b, 'gray', linestyle='--', linewidth=1)

    ax.set_xlabel('Nombre de sens (WordNet)')
    ax.set_ylabel('Score BERT')
    ax.set_title(f'{titre}\nSpearman ρ={rho:.3f}, p={p:.4f}')
    ax.legend(handles=[
        mpatches.Patch(color='#E85D24', label='polysémique'),
        mpatches.Patch(color='#1D9E75', label='monosémique')
    ])
    print(f'[{lang.upper()}] Spearman ρ = {rho:.3f}, p = {p:.4f}')

plt.tight_layout()
plt.savefig('correlation_fr_en.png', dpi=150)
plt.show()
print('✅ Sauvegardé : correlation_fr_en.png')
```

```python
# ============================================================
# CELLULE 12 — t-SNE GLOBAL : FR et EN
# ============================================================

def tsne_global(embeddings, mots_poly, titre, fichier):
    mots_uniques = list(embeddings.keys())
    all_vecs, all_mots = [], []
    for mot in mots_uniques:
        for vec in embeddings[mot]:
            all_vecs.append(vec)
            all_mots.append(mot)

    X = np.array(all_vecs)
    X_tsne = TSNE(n_components=2, perplexity=30,
                  random_state=42, max_iter=1000).fit_transform(X)

    palette = plt.cm.tab20.colors
    coul = {m: palette[i % 20] for i, m in enumerate(mots_uniques)}

    fig, ax = plt.subplots(figsize=(11, 7))
    for mot in mots_uniques:
        idx = [i for i, m in enumerate(all_mots) if m == mot]
        coords = X_tsne[idx]
        marker = 'o' if mot in mots_poly else '^'
        ax.scatter(coords[:, 0], coords[:, 1],
                   color=coul[mot], marker=marker, alpha=0.7, s=35, label=mot)
        ax.annotate(mot, (coords[:, 0].mean(), coords[:, 1].mean()),
                    fontsize=8, fontweight='bold', color=coul[mot],
                    bbox=dict(boxstyle='round,pad=0.2', fc='white', alpha=0.6))

    ax.set_title(f't-SNE — {titre} (● polysémique  ▲ monosémique)')
    ax.legend(bbox_to_anchor=(1.01, 1), loc='upper left', fontsize=7, ncol=1)
    plt.tight_layout()
    plt.savefig(fichier, dpi=150, bbox_inches='tight')
    plt.show()
    print(f'✅ Sauvegardé : {fichier}')

tsne_global(embeddings_fr, mots_poly_fr, 'CamemBERT Français', 'tsne_fr.png')
tsne_global(embeddings_en, mots_poly_en, 'BERT-base Anglais',  'tsne_en.png')
```

```python
# ============================================================
# CELLULE 13 — PCA ISOLÉE : mots les plus intéressants
# ============================================================

def pca_mot(mot, embeddings, phrases_retenues, mots_poly, langue, n_exemples=3):
    vecs = np.array(embeddings[mot])
    if len(vecs) < 4:
        print(f'⚠️  Pas assez de vecteurs pour {mot}')
        return
    pca = PCA(n_components=2)
    coords = pca.fit_transform(vecs)
    color = '#E85D24' if mot in mots_poly else '#1D9E75'

    fig, ax = plt.subplots(figsize=(6, 5))
    ax.scatter(coords[:, 0], coords[:, 1], alpha=0.7, s=60, color=color)

    for i in range(min(n_exemples, len(coords))):
        phrase_courte = phrases_retenues[mot][i][:40] + '...'
        ax.annotate(phrase_courte, (coords[i, 0], coords[i, 1]),
                    fontsize=6, alpha=0.6)

    var = pca.explained_variance_ratio_
    ax.set_title(f'[{langue.upper()}] « {mot} » — PCA 2D\n'
                 f'(variance : {var[0]:.0%} + {var[1]:.0%})')
    ax.set_xlabel('PC1')
    ax.set_ylabel('PC2')
    plt.tight_layout()
    plt.savefig(f'pca_{langue}_{mot}.png', dpi=150)
    plt.show()

# Mots les plus intéressants FR
for mot in ['tour', 'avocat', 'souris', 'chaise']:
    pca_mot(mot, embeddings_fr, phrases_fr, mots_poly_fr, 'fr')

# Mots les plus intéressants EN
for mot in ['bank', 'mouse', 'crane', 'chair']:
    pca_mot(mot, embeddings_en, phrases_en, mots_poly_en, 'en')
```

```python
# ============================================================
# CELLULE 14 — TABLEAU DE COMPARAISON FINALE
# ============================================================
df_fr['langue'] = 'FR'
df_en['langue'] = 'EN'
df_all = pd.concat([df_fr, df_en], ignore_index=True)

# Statistiques par langue et type
resume = df_all.groupby(['langue', 'type'])['score_bert'].agg(
    moyenne='mean', ecart_type='std'
).round(4)
print('\n=== Résumé statistique ===')
print(resume.to_string())

# Corrélations finales
print('\n=== Corrélations Spearman ===')
for langue, df in [('FR', df_fr), ('EN', df_en)]:
    rho, p = spearmanr(df['nb_sens'], df['score_bert'])
    print(f'  [{langue}] ρ = {rho:.3f}, p = {p:.4f}')
```
